```python
from main import DistanceList1
from main import DistanceList2
from statistics import mean
import pandas as pd
import csv

number_of_products1 = []
number_of_orders1 = []
avg_distance1 = []
for N in range(100,400,20):
    for M in range(100,200,25):
        number_of_products1.append(N)
        number_of_orders1.append(M)
        avg_distance1.append(mean(DistanceList1(N,M)))

dictionary1 = {"N": number_of_products1, "M": number_of_orders1, "Average Distance": avg_distance1}
data1 = pd.DataFrame.from_dict(dictionary1)

with open('data1.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(data1)
    
number_of_products2 = []
number_of_orders2 = []
avg_distance2 = []
for N in range(100,400,20):
    for M in range(100,200,25):
        number_of_products2.append(N)
        number_of_orders2.append(M)
        avg_distance2.append(mean(DistanceList2(N,M)))

dictionary2 = {"N": number_of_products2, "M": number_of_orders2, "Average Distance": avg_distance2}
data2 = pd.DataFrame.from_dict(dictionary2)

with open('data2.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(data2)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Select the data to plot
x = data1['N'].to_numpy()
y = data1['M'].to_numpy()
z = np.zeros(len(x))
dx = np.ones(len(x))
dy = np.ones(len(x))
dz = data1["Average Distance"].to_numpy()

# Create a figure and an axes
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D bar chart
ax.bar3d(x, y, z, dx, dy, dz, color="grey")

# Set the x, y, and z labels
ax.set_xlabel('Number of Products (N)')
ax.set_ylabel('Orders (M)')
ax.set_zlabel('Average Distance')


# Show the plot
plt.title("Distribution of Distances with Sorting")
plt.show()
```
